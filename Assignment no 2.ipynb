```python
#43. Digitalize the book allotment process for school. Charges are mentioned here in the given table:
# Book Prices Dictionary
Book_price = {
    '1st':{'Hindi':60, 'Marathi':60, 'English':80, 'Science':90, 'Maths':100},

    '2nd':{'Hindi':60, 'Marathi':60, 'English':80, 'Science':90, 'Maths':100},

    '3rd':{'Hindi':60, 'Marathi':60, 'English':80, 'Science':90, 'Maths':100},

    '4th':{'Hindi':60, 'Marathi':60, 'English':80, 'Science':90, 'Maths':100},
 
    '5th':{'Hindi':100, 'Marathi':100, 'English':100, 'Science':120, 'Maths':140},

    '6th':{'Hindi':100, 'Marathi':100, 'English':100, 'Science':120, 'Maths':140},

    '7th':{'Hindi':100, 'Marathi':100, 'English':100, 'Science':120, 'Maths':140},

    '8th':{'Hindi':100, 'Marathi':100, 'English':100, 'Science':120, 'Maths':140},
 
    '9th':{'Hindi':150, 'Marathi':150, 'English':150, 'Science':200, 'Maths':250},

    '10th':{'Hindi':150, 'Marathi':150, 'English':150, 'Science':200, 'Maths':250}
}
 
# Notebook Prices Dictionary
Notebook_price = {
    'square':{'100 pages':40, '200 pages':70},

    '4lines':{'100 pages':30, '200 pages':50},

    '2lines':{'100 pages':30, '200 pages':50},

    'single lines':{'100 pages':60, '200 pages':100},

    'A4 Notebook':{'100 pages':100, '200 pages':180}
}
def CalBook(standard):  #class enter by user
    book_total = 0
    if standard not in Book_price:
        print("Invalid class selected")
        return 0

    print("\nAvailable Books:")
    for book, price in Book_price[standard].items():
        print(book, "-", price)

    while True:
        subject = input("Enter book name (or 'stop'): ").lower()

        if subject == "stop":
            break

        if subject in Book_price[standard]:
            book_total += Book_price[standard][subject]
            print(subject, "added")
        else:
            print("Book not available")

    return book_total

def CalNoteBook():
    notebook_total = 0
    print("\nAvailable Notebooks:")
    for n_type in Notebook_price:
        print(n_type)

    while True:
         n_type = input("Enter notebook type (or 'stop'): ")

         if n_type.lower() == "stop":
            break

         if n_type in Notebook_price:
            pages = input("Enter pages (100 pages / 200 pages): ")

            if pages in Notebook_price[n_type]:
                notebook_total += Notebook_price[n_type][pages]
            else:
                print("Invalid page option")
         else:
            print("Notebook type not available")

    return notebook_total

name = input("Enter student name:")
standard = input("Enter class (1st to 10th):")
choice = input("Do you want book/ notebook / both:").lower()

book_total = 0
notebook_total = 0

if choice == "book":
    book_total = Book_price(standard)

elif choice == "notebook":
    notebook_total = CalNoteBook()

elif choice == "both":
    book_total = CalBook(standard)
    notebook_total = CalNoteBook()
else:
    print("Thank you for visiting")

Final_amount = book_total + notebook_total
print("\n------ BILL SUMMARY ------")
print("Student Name:", name)
print("Class:",standard)
print("Books Cost:",book_total)
print("Notebooks Cost:",notebook_total)
print("Final Amount:",Final_amount)
```

```python
# 49.Create a function to calculate age till now.
from datetime import datetime
def CalAge(dob):
   
    dob_date = datetime.strptime(dob, "%d-%m-%Y").date()
    today = datetime.today().date()
    age = today.year - dob_date.year
    if (today.month, today.day) < (dob_date.month, dob_date.day):
        age =age- 1
    return age
dob= input("Enter a DOB")  
CalAge(dob)
```
